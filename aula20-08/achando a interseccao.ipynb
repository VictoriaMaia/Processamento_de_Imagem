```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
im = cv2.imread('test1.jpg')
img1 = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

im = cv2.imread('test2.jpg')
img2 = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

plt.figure(figsize=[15,15])
plt.subplot(121), plt.imshow(img1)
plt.subplot(122), plt.imshow(img2)
plt.show()
```

```python
print(" imagem1", img1.shape, "\n", "imagem2", img2.shape)
```

```python
xIm1 = img1.shape[0]
yIm1 = img1.shape[1]
xIm2 = img2.shape[0]
yIm2 = img2.shape[1]
```

```python
def mse(xBegin, xEnd, yBegin, yEnd):
    #print(xBegin, xEnd, yBegin, yEnd, '|||', (xEnd-xBegin), (yEnd-yBegin))
    return np.mean((img1[xBegin : xEnd , yBegin : yEnd] - img2[0 : (xEnd-xBegin) , 0 : (yEnd-yBegin)]) ** 2)
```
