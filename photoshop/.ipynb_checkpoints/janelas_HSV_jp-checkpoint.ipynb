```python
import sys
import numpy as np
import cv2
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d



def callback_func(x):
    pass


cv2.namedWindow('image',cv2.WINDOW_NORMAL)
# cv2.SetWindowProperty('image', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN);



im = cv2.imread("panda.jpg")
hsv_im = cv2.cvtColor(im, cv2.COLOR_BGR2HSV)
hsv_im2 = hsv_im
# h_median = int(np.median(hsv_im[:,:,0]))
# s_median = int(np.median(hsv_im[:,:,1]))
# v_median = int(np.median(hsv_im[:,:,2]))

# width = im.shape[1]
# height = im.shape[0]
# print(hsv_mean.shape)

# h, s, v = hsv_im[:, :, 0], hsv_img[:, :, 1], hsv_img[:, :, 2]
aux =  np.zeros((1080,1440))
hue = 0
hue_old = 0
sat = 0
sat_old = 0
val = 0
val_old = 0


cv2.createTrackbar('matiz','image',0,179,callback_func)

cv2.createTrackbar('saturacao','image',0,255,callback_func)
cv2.createTrackbar('valor','image',0,255,callback_func)

while(1):    
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break
    
    hue = cv2.getTrackbarPos('matiz','image')
    if hue != hue_old:
        aux.fill(hue - hue_old)
        hsv_im2[:,:,0] = hsv_im[:,:,0] + aux
#         print(hsv_im[:,:,0])
        hue_old = hue
        
    sat = cv2.getTrackbarPos('saturacao','image')
    if sat != sat_old:
        aux.fill(sat - sat_old)
        hsv_im2[:,:,1] = cv2.add(hsv_im[:,:,1], aux, dtype=cv2.CV_8U)
        sat_old = sat
        
    val = cv2.getTrackbarPos('valor','image')
    if val != val_old:
        aux.fill(val - val_old)
        hsv_im2[:,:,2] = cv2.add(hsv_im[:,:,2], aux, dtype=cv2.CV_8U)
        val_old = val
    
    
    im2 = cv2.cvtColor(hsv_im2, cv2.COLOR_HSV2BGR)
    cv2.imshow('image',im2)


cv2.destroyAllWindows()
```
