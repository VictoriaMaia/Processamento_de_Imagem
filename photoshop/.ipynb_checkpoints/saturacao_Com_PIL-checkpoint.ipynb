```python
import cv2
import numpy
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("panda.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image);
```

```python
plt.figure(figsize=[10,10])
plt.subplot(312), plt.hist(image.ravel(), 256, [0,256]);
```

```python
img = image.copy()
img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
```

```python
plt.figure(figsize=[10,10])
plt.subplot(312), plt.hist(img[:,:,1].ravel(), 256, [0,256]);
```

```python
plt.figure(figsize=[10,10])
plt.subplot(312), plt.hist(img[:,:,1].ravel(), 256, [0,256]);
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
```

```python
img[:,:,1] = cv2.equalizeHist(img[:,:,1])
```

```python
# img = image.copy()
valor = 200
```

```python
# for i in range(img.shape[2]):
for j in range(img.shape[1]):
    for k in range(img.shape[0]):
#             if img[k,j,i] < valor:
            if img[k,j,1] + valor > 255:
                img[k,j,1] = 255
            else:
                img[k,j,1] = img[k,j,1] + valor
```

```python
img[:,:,1]
```

```python
img[:,:,1] = map(lambda x: x+valor > 255 or x+valor, img[:,:,1])
```

```python
img[:,:,1]
```

```python
img[:,:,0] = cv2.equalizeHist(img[:,:,0])
img[:,:,1] = cv2.equalizeHist(img[:,:,1])
img[:,:,2] = cv2.equalizeHist(img[:,:,2])
```

```python
img = cv2.cvtColor(img, cv2.COLOR_HSV2RGB)
```

```python
plt.figure(figsize=[10,10])
plt.subplot(312), plt.hist(img.ravel(), 256, [0,256]);
```

```python
img = cv2.normalize(img, None, 0, 256, cv2.NORM_MINMAX, cv2.CV_8U)
```

```python
plt.figure(figsize=[25,25])
plt.subplot(121), plt.imshow(image)
plt.subplot(122), plt.imshow(img)
plt.show()
```
