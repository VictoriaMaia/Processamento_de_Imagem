bolinhas = 228

```python
_, bin_bolinhas = cv2.threshold(bolinhas, 228, 255, cv2.THRESH_BINARY_INV)
plt.imshow(bin_bolinhas, 'gray');
```

```python
elem_type_rect = cv2.MORPH_RECT
element_rect = cv2.getStructuringElement(elem_type_rect, (3, 3))
```

```python
erosion_bolinhas = cv2.erode(bin_bolinhas, element_rect)
plt.imshow(erosion_bolinhas, 'gray');
```

```python
for i in range(0,11):
    erosion_bolinhas = cv2.erode(erosion_bolinhas, element_rect)
plt.imshow(erosion_bolinhas, 'gray');
```

```python
dilation_bolinhas = cv2.dilate(erosion_bolinhas, element_rect)
plt.imshow(dilation_bolinhas, 'gray');
```

```python
for i in range(0, 34):
    dilation_bolinhas = cv2.dilate(dilation_bolinhas, element_rect)
plt.imshow(dilation_bolinhas, 'gray');
```

```python
final_bolinhas = cv2.bitwise_and(bin_bolinhas, dilation_bolinhas)
plt.imshow(final_bolinhas, 'gray');
```
