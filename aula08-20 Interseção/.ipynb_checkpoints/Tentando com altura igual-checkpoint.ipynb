```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
img1 = cv2.imread('part1.jpg', 0)
# img1 = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

img2 = cv2.imread('part2.jpg', 0)
# img2 = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

# plt.figure(figsize=[15,15])
plt.subplot(121), plt.imshow(img1, 'gray')
plt.subplot(122), plt.imshow(img2, 'gray')
plt.show()
```

```python
print("1: ", img1.shape, "\n2: ", img2.shape)
```

```python
mse = mean_squared_error
```

```python
mse(img1, img1)
```

```python
plt.subplot(121), plt.imshow(img1[:, -100:], 'gray')
plt.subplot(122), plt.imshow(img2[:, 296:396], 'gray')
# plt.subplot(122), plt.imshow(img2, 'gray')
plt.show()
```

```python
mse(img1[:,-100:], img2[:,296:396])
```

```python
for i in range(0, 481, 1):
    res = mse(img1[:, -100:], img2[:, i:(i+100)])

    if res < 5:
        print('Deu bom no i = '+str(i))
        break
```
