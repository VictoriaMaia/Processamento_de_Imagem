```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Lendo imagem e convertendo para RGB

im = cv2.imread('timao_e_pumba.jpg')
im_original = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

plt.imshow(im_original);
```

```python
# Scaling - muda o tamanho
# Diminuir 20%

M_scaling = np.float32([[.8,0,1], [0,.8,1]])

# Dimenses da imagem

width = im_original.shape[1]
height = im_original.shape[0]

# Aplicar o scaling na imagem.

im_scaled = cv2.warpAffine(im_original,M_scaling,(width,height), borderValue=(255,255,255))

plt.imshow(im_scaled);
im_scaled = cv2.resize(im_scaled,(600,350))
im_scaled.shape
```

```python
im_T = cv2.resize(im_scaled,(600,350))
plt.imshow(im_T);
im_T.shape
```

```python
# Rotacionando
# M_rotation = cv2.getRotationMatrix2D((x_center,y_center),angulacao,1)

M_rotation30 = cv2.getRotationMatrix2D((0,0), 30, 1)
M_rotation45 = cv2.getRotationMatrix2D((0,0), 45, 1)
M_rotation_por_ai = cv2.getRotationMatrix2D((250,100), 30, 1)

im_rotated30 = cv2.warpAffine(im_original,M_rotation30,(width,height))
im_rotated45 = cv2.warpAffine(im_original,M_rotation45,(width,height))
im_rotated_por_ai = cv2.warpAffine(im_original,M_rotation_por_ai,(width,height))

plt.figure(figsize=[15,15])
plt.subplot(131), plt.imshow(im_rotated30)
plt.subplot(132), plt.imshow(im_rotated45)
plt.subplot(133), plt.imshow(im_rotated_por_ai)
plt.show()
```

```python
# Transladar (x,y) = (300,100)

M_translation = np.float32([[1,0,300],[0,1,100]])
im_translated = cv2.warpAffine(im_original,M_translation,(width,height))

plt.imshow(im_translated);
```
