```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('panterinha.jpg',0)
plt.imshow(img, 'gray');
```

```python
# Filtro media 5x5
avg_blur = cv2.blur(img,(5,5))
plt.imshow(avg_blur, 'gray');
```

```python
# Gaussian
gaussian_blur = cv2.GaussianBlur(img,(5,5),1)
plt.imshow(gaussian_blur, 'gray');
```

```python
# mediana
median_blur = cv2.medianBlur(img,5)
plt.imshow(median_blur, 'gray');
```

```python
# bilateral
# aumentando o valor fica parecido com catoon
bilateral_blur = cv2.bilateralFilter(img,9,100,100)
plt.imshow(bilateral_blur, 'gray');
```

```python
# custom kernel (media 3x3)
kernel_media = np.float32([[1,1,1],[1,1,1],[1,1,1]])/9
customMedia = cv2.filter2D(img,-1,kernel_media)

# custom realcar as bordas (media - imagem)
plt.figure(figsize=[10,10])
customBordas = cv2.subtract(img, customMedia)
# customBordas = cv2.bitwise_not(customBordas)
plt.imshow(customBordas, 'gray');
```

```python
# custom realcar as bordas na imagem (img + customBordas)
plt.figure(figsize=[10,10])
custom = cv2.add(img, customBordas)
plt.imshow(custom, 'gray');
```

```python
plt.figure(figsize=[20,20])
plt.subplot(231),plt.imshow(img, 'gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(232),plt.imshow(avg_blur, 'gray'),plt.title('Media')
plt.xticks([]), plt.yticks([])
plt.subplot(233),plt.imshow(gaussian_blur, 'gray'),plt.title('Gaussian')
plt.xticks([]), plt.yticks([])
plt.subplot(234),plt.imshow(median_blur, 'gray'),plt.title('Mediana')
plt.xticks([]), plt.yticks([])
plt.subplot(235),plt.imshow(bilateral_blur, 'gray'),plt.title('Bilateral')
plt.xticks([]), plt.yticks([])
plt.subplot(236),plt.imshow(custom, 'gray'),plt.title('Custom')
plt.xticks([]), plt.yticks([])
plt.show()
```
