```python
import sys
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Abrir uma imagem
# imread(nome_arquivo, label) sendo 0 para abrir a imagem em tons de cinza

im = cv2.imread('einstein.jpg', 0) 
plt.imshow(im, 'gray');
```

```python
# Tamanho da imagem

print (im.shape)
```

```python
# Redimensionando a imagem

width = int(im.shape[1] * .5)
height = int(im.shape[0] * .5)
dim = (width, height)

im_resized = cv2.resize(im, dim, interpolation = cv2.INTER_AREA)
```

```python
# Calculando o histograma da imagem (continuo)

im_hist = cv2.calcHist([im], [0], None, [256], [0,256])
plt.figure(figsize=[10,10])
# plt.subplot(211), plt.imshow(im,'gray')
plt.subplot(212), plt.plot(im_hist)
plt.show()
```

```python
# Outra forma de calcular o histograma (discreto)

plt.figure(figsize=[10,10])
plt.subplot(312), plt.hist(im.ravel(), 256, [0,256]);
```

```python
# Equalizar o histograma

im_eq_cdf = cv2.equalizeHist(im)

plt.figure(figsize=[10,10])
plt.subplot(321), plt.imshow(im,'gray')
plt.subplot(322), plt.imshow(im_eq_cdf,'gray')
plt.subplot(323), plt.hist(im.ravel(), 256, [0,256])
plt.subplot(324), plt.hist(im_eq_cdf.ravel(), 256, [0,256])
plt.show()
```
