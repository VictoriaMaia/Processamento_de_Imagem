```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Lendo imagem e convertendo para RGB

im = cv2.imread('vulcao.jpeg')
imgVulc = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

im = cv2.imread('pessoa.jpg')
imgPess = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

plt.subplot(121), plt.imshow(imgVulc)
plt.subplot(122), plt.imshow(imgPess)
plt.show()
```

```python
# Tamanho das imagens

print("V: ", imgVulc.shape, "P: ", imgPess.shape)
```

```python
# Redimensionar

dim = (imgPess.shape[1], imgPess.shape[0]) #conjunto de dimensões
vulc = cv2.resize(imgVulc, dim) #resize(imagem, dimensoes_novas)

print("V: ", vulc.shape, "P: ", imgPess.shape)
plt.imshow(vulc);
```

```python
# Separando os canais da imagem pessoa

imgP_b, imgP_g, imgP_r = cv2.split(imgPess)

plt.figure(figsize=[15,15])
plt.subplot(141), plt.imshow(imgP_b, 'gray')
plt.subplot(142), plt.imshow(imgP_g, 'gray')
plt.subplot(143), plt.imshow(imgP_r, 'gray')
plt.subplot(144), plt.imshow(imgPess)
plt.show()
```

```python
# Histograma apenas do canal verde 
# Para poder pegar o valor do verde para retirar o fundo.

plt.hist(imgP_g.ravel(), 256, [0,256]);
```

```python
# Fazendo Thresold
# Cria uma mascara onde o branco é 1 e o preto é 0 utilizando apenas o canal verde que é a cor 
# que a gente quer tirar

ret,im_threshPessoa = cv2.threshold(imgP_g, 210, 255, cv2.THRESH_BINARY)
plt.imshow(im_threshPessoa, 'gray');
```

```python
# b (2)
vulcCopy = vulc.copy()
vulcCopy[:, :, 0] = 0
vulcCopy[:, :, 1] = 0
plt.imshow(vulcCopy);

# g (1)
vulcCopy = vulc.copy()
vulcCopy[:, :, 0] = 0
vulcCopy[:, :, 2] = 0
plt.imshow(vulcCopy);

# r (0)
vulcCopy = vulc.copy()
vulcCopy[:, :, 1] = 0
vulcCopy[:, :, 2] = 0
plt.imshow(vulcCopy);

# g e b
vulcCopy = vulc.copy()
vulcCopy[:, :, 0] = 0
plt.imshow(vulcCopy);

# r e b
vulcCopy = vulc.copy()
vulcCopy[:, :, 1] = 0
plt.imshow(vulcCopy);

# g e r
vulcCopy = vulc.copy()
vulcCopy[:, :, 2] = 0
plt.imshow(vulcCopy);
```

```python
# Separando os canais da imagem vulcão

imgV_b, imgV_g, imgV_r = cv2.split(vulc)

plt.figure(figsize=[15,15])
plt.subplot(141), plt.imshow(imgV_b)
plt.subplot(142), plt.imshow(imgV_g)
plt.subplot(143), plt.imshow(imgV_r)
plt.subplot(144), plt.imshow(vulc)
plt.show()
```

```python
# Criando uma máscara de 3 canais

mascara3C = cv2.merge((im_threshPessoa, im_threshPessoa, im_threshPessoa))
```

```python
# Ajeitando o fundo para ficar apenas o vulcão
# Fazendo uma AND da mascara com a imagem de vulcão para deixar o fundo do vulcão
# Como o branco (representando o fundo) é 1, então vulcão AND 1 = vulcão.

imagemFinalVulc = cv2.bitwise_and(mascara3C, vulc)
plt.imshow(imagemFinalVulc, 'gray');
```

```python
# Invertendo mascara3C para que agora a mulher fique branca(1) e o fundo preto(0)

mascaraInversa3C = cv2.bitwise_not(mascara3C)
plt.imshow(mascaraInversa3C, 'gray');
```

```python
# Ajeitando para ficar apenas a mulher no meio
# Fazendo uma AND da máscara com a mulher, vai ficar só a mulher
# mulher AND 1 = mulher

imagemFinalPes = cv2.bitwise_and(mascaraInversa3C, imgPess)
plt.imshow(imagemFinalPes, 'gray');
```

```python
# Agora que tem uma imagem final apenas com o fundo e uma imagem final apenas com a mulher
# Precisa apenas de uma soma ou um add para juntas as duas imagens

imagemFinalzona = cv2.add(imagemFinalVulc, imagemFinalPes)
plt.figure(figsize=[15,15])
plt.imshow(imagemFinalzona, 'gray');
```
